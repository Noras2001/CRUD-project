### **<span style="color: orange;">Домашнее задание: Создание функционального приложения с расширенными CRUD-операциями и агрегатными запросами</span>**

**<span style="color: green;">Цель</span>** — Разработать мини-приложение для управления каталогом товаров с возможностью фильтрации, сортировки и агрегирования данных. Задание направлено на отработку сложных CRUD-операций, а также на углубленное понимание работы с Django ORM, включая аннотации и агрегаты.

#### <span style="color: orange;">Описание задания</span>

Вам предстоит создать приложение для управления каталогом товаров с поддержкой следующих функций:

1. **<span style="color: yellow;">Создание и редактирование товаров</span>** с полями:
   - <span style="color: green;">name</span> (название товара)
   - <span style="color: green;">price</span> (цена)
   - <span style="color: green;">description</span> (описание товара)
   - <span style="color: green;">created_at</span> (дата добавления)
   - <span style="color: green;">updated_at</span> (дата обновления)

2. **<span style="color: yellow;">Работа с категориями</span>**:
   - Создать модель <span style="color: green;">Category</span> с полем <span style="color: green;">name</span>.
   - Связать товары с категориями через ForeignKey.
   - Добавить возможность управлять категориями (CRUD) и привязывать к ним товары.

3. **<span style="color: yellow;">Расширенный CRUD для товаров</span>**:
   - Реализовать создание, редактирование и удаление товаров через интерфейс с использованием форм.
   - Сделать поле <span style="color: green;">price</span> обязательным и добавлять валидацию для цен ниже нуля.

4. **<span style="color: yellow;">Фильтрация и сортировка</span>**:
   - Добавить возможность фильтрации товаров по категории и по диапазону цен.
   - Реализовать сортировку по цене и дате добавления (в порядке убывания и возрастания).

5. **<span style="color: yellow;">Агрегатные функции и аналитика</span>**:
   - Подсчитать общую стоимость всех товаров и количество товаров в каждой категории.
   - Вывести среднюю цену товаров и минимальную/максимальную цену для каждой категории.
   - Реализовать представление, где будет отображаться список категорий с общей стоимостью товаров внутри каждой.

6. **<span style="color: yellow;">Дополнительные требования</span>**:
   - Создать подтверждение перед удалением товара.
   - Реализовать страницу с аналитикой, где отображаются итоговые значения по категориям и общая статистика каталога (например, средняя цена, общее количество товаров, самая высокая и самая низкая цена).

#### <span style="color: orange;">Технические требования</span>

- **<span style="color: yellow;">Структура приложения</span>**: <span style="color: green;">catalog</span> (папка приложения), в которой создаются модели, представления, шаблоны и маршруты.
- **<span style="color: yellow;">Шаблоны</span>**: Добавить страницы для создания и редактирования товара, категорий, и просмотра аналитики. Шаблоны должны быть оформлены с использованием Bootstrap или другого CSS-фреймворка для улучшения пользовательского интерфейса.
- **<span style="color: yellow;">Код</span>**: Написать качественно документированный код с комментариями к каждому сложному методу или функции.

#### <span style="color: orange;">Критерии оценки</span>

- Корректность выполнения CRUD-операций.
- Реализация фильтрации, сортировки и агрегатных запросов.
- Использование Django ORM для выполнения всех запросов, без использования «сырых» SQL.
- Удобство и функциональность интерфейса.
- Качество кода и структурированность приложения.
